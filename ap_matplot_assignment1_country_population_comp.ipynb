```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#reading data from .csv file

x=pd.read_csv("countries.csv")
x
```

```python
#read the required country data from the above x data

ch=x[x.country=='China']
ch

ind=x[x.country=='India']
ind

us=x[x.country=='United States']
us

uk=x[x.country=='United Kingdom']
uk

ger=x[x.country=='Germany']
ger
```

```python
#plot the population vs year curves for all the countries and labelling the axis,legend and curves
plt.plot(ch.year,ch.population)
plt.plot(ind.year,ind.population)
plt.plot(us.year,us.population)
plt.plot(uk.year,uk.population)
plt.plot(ger.year,ger.population)

plt.title('Country wise Population comparison')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(['China','India','United States','United Kingdom','Germany'])
plt.show()
```
